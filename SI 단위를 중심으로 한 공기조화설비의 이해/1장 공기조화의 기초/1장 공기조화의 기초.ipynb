## **1장 연습문제**

### **Problem 01**  


최근 우리의 생활수준이 선진화되면서부터 대형 건물은 물론이고, 가정에서도 형편만 된다면 소형 에어컨을 설치하려고 하고 있다. 그러면 우리들에게 에어컨이라는 기계로 통칭되는 공기조화기(air-conditioner)란 무엇을 하는 기계이며, 일상생활에서 공기조화(air-conditioning)가 왜 필요한지에 대하여 생각해 보자.

#### **Answer**


공기조화기는 단순히 공기를 차갑게 만드는 기계가 아니라, 실내의 공기를 인간이 쾌적하게 느끼도록 온도, 습도, 청정도 및 기류를 적절한 상태로 유지·조절하는 장치이다. 즉, 공기조화는 냉방뿐만 아니라 난방, 가습, 제습, 환기 및 공기정화 기능까지 포함하는 종합적인 실내 환경 조절 기술이다.

공기조화기의 주요 기능은 다음과 같다.

온도 조절
여름철에는 실내의 열을 제거하여 냉방하고, 겨울철에는 열을 공급하여 난방함으로써 인간이 쾌적하게 생활할 수 있는 온도를 유지한다.
습도 조절
습도가 너무 높으면 불쾌감을 느끼고 곰팡이 발생이 쉬우며, 너무 낮으면 피부 건조와 호흡기 질환이 발생하기 쉽다. 공기조화기는 가습 및 제습 기능을 통해 적절한 습도를 유지한다.
공기 청정
먼지, 세균, 유해가스 등을 제거하여 깨끗한 공기를 공급한다. 특히 현대 도시 환경에서는 미세먼지와 오염물질 제거가 매우 중요하다.
공기 순환 및 환기
실내 공기를 순환시키고 외부의 신선한 공기를 도입하여 산소 농도를 유지하고 이산화탄소 및 오염물질을 제거한다.

현대 생활에서 공기조화가 필요한 이유는 다음과 같다.

첫째, 인간의 쾌적성과 건강 유지를 위해 필요하다. 인간은 온도와 습도가 적절한 환경에서 가장 쾌적함을 느끼며 작업 능률도 향상된다. 지나치게 덥거나 추운 환경은 피로와 스트레스를 유발하고 건강에도 악영향을 미친다.

둘째, 산업 및 정보화 사회에서 필수적이다. 병원, 반도체 공장, 데이터센터, 연구실 등은 일정한 온·습도와 청정 환경 유지가 매우 중요하므로 공기조화 설비가 반드시 필요하다.

셋째, 건축물의 대형화·고층화 때문이다. 현대 건물은 밀폐 구조가 많아 자연 환기만으로는 쾌적한 실내 환경을 유지하기 어렵기 때문에 기계적인 공기조화가 필요하다.

넷째, 생활수준 향상에 따른 쾌적한 주거환경 요구 증가 때문이다. 경제 발전과 함께 사람들은 단순한 생존이 아니라 보다 쾌적하고 건강한 생활환경을 추구하게 되었으며, 이에 따라 가정용 에어컨 보급도 확대되고 있다.

따라서 공기조화는 인간의 건강과 쾌적한 생활, 작업 능률 향상 및 산업 활동 유지에 필수적인 기술이라고 할 수 있다.

### **Problem 02**  

기계를 이용하지 않고 자연의 힘을 이용하여 공기조화하는 방법으로는 어떤 것이 있는지 생각해 보고, 국내·외에서 자연공조에 해당하는 예를 찾아보자.

#### **Answer**

자연의 힘을 이용한 공기조화란 기계식 냉동기나 송풍기를 사용하지 않고 바람, 태양열, 물, 지열 등 자연에너지를 이용하여 실내 환경을 쾌적하게 유지하는 방법을 말한다. 이를 자연공조 또는 패시브 공조라고 한다.  


자연공조의 방법에는 다음과 같은 것들이 있다.

1. **자연환기**  
창문, 환기구, 중정 등을 이용하여 바람이 자연스럽게 실내를 통과하도록 하는 방법이다. 풍압과 온도차에 의해 공기가 이동하며 실내의 더운 공기와 오염된 공기를 배출한다.  
2. **굴뚝효과 이용**  
더운 공기는 위로 상승하는 성질이 있으므로 건물 상부에 배기구를 설치하여 공기의 자연 순환을 유도하는 방법이다. 아트리움이나 환기탑에 많이 이용된다.  
3. **일사 및 차양 이용**  
겨울에는 태양열을 받아 난방 효과를 얻고, 여름에는 차양장치나 처마를 이용하여 직사광선을 차단함으로써 냉방부하를 줄인다.  
4. **지열 및 토양 이용**  
지중의 온도가 연중 거의 일정한 점을 이용하여 외기를 지하 공간이나 지중관을 통과시켜 여름에는 냉각하고 겨울에는 예열한다.  
5. **수공간 및 증발냉각 이용**  
분수, 연못, 식물 등을 이용하여 물의 증발 시 발생하는 냉각효과로 주변 온도를 낮추는 방법이다.  
6. **녹화 및 단열 구조 활용**  
옥상녹화, 벽면녹화 및 고단열 구조를 통해 외부 열의 유입을 줄이고 실내 온도를 안정적으로 유지한다.    

국내의 자연공조 사례로는 다음과 같은 예가 있다.

1. **한옥**  
대청마루와 창호 구조를 이용하여 여름철에는 통풍이 잘 되도록 하고, 겨울철에는 온돌을 이용하여 난방을 하였다. 이는 대표적인 한국의 전통 자연공조 방식이다.  
2. **서울시 신청사**  
이중외피와 자연환기 시스템을 적용하여 에너지 사용을 줄이고 자연채광 및 자연통풍을 적극 활용하였다.  
3. **제주도의 전통가옥**
강한 바람과 지역 기후를 고려하여 배치와 창 구조를 설계하여 자연환기를 극대화하였다.  

국외의 자연공조 사례는 다음과 같다.  

1. **독일 국회의사당(Reichstag)**  
유리 돔 구조와 자연환기 시스템을 이용하여 내부 공기를 순환시키고 채광을 확보하였다.  
2. **영국 런던의 거킨 빌딩(30 St Mary Axe)**  
건물 외피 사이에 환기 공간을 두어 자연통풍이 가능하도록 설계하였다.  
3. **중동 지역의 바드기르(Badgir, 풍탑) 건축**  
사막지역의 뜨거운 공기를 냉각시키기 위해 풍탑을 설치하여 자연 바람을 실내로 유도하였다.  

이처럼 자연공조는 에너지 소비를 줄이고 친환경적인 실내 환경을 조성할 수 있기 때문에 현대 건축에서도 중요성이 점점 커지고 있다.

### **Problem 03**  

공기조화의 기본 개념을 생각하여 기계 사용을 최소화하면서 공기조화를 할 수 있는 방법에 대해 생각해 보자.

#### **Answer**

공기조화의 기본 목적은 인간이 쾌적함을 느낄 수 있도록 실내의 온도, 습도, 기류 및 공기청정 상태를 적절하게 유지하는 데 있다. 그러나 기계식 냉난방 장치는 많은 에너지를 소비하므로, 기계 사용을 최소화하면서 자연의 힘을 적극 활용하는 공기조화 방식이 필요하다.  

기계 사용을 최소화하면서 공기조화를 할 수 있는 방법에는 다음과 같은 것들이 있다.

1. **자연환기 적극 활용**  
건물의 창문과 환기구를 적절히 배치하여 외부의 바람이 실내를 자연스럽게 통과하도록 한다. 특히 맞통풍 구조를 적용하면 실내의 더운 공기와 오염된 공기를 효과적으로 배출할 수 있다.  

2. **단열 성능 향상**  
벽체, 지붕, 창문의 단열 성능을 높이면 외부 열의 유입과 실내 열 손실을 줄일 수 있다. 이를 통해 냉방기와 난방기의 사용을 줄일 수 있다.  

3. **차양 및 일사 조절**  
처마, 블라인드, 루버 등을 이용하여 여름철에는 강한 태양복사를 차단하고, 겨울철에는 태양열을 실내로 유입시켜 냉난방 부하를 줄인다.  

4. **건물 배치와 방향 고려**  
건물을 설계할 때 태양의 이동 경로와 바람 방향을 고려하면 자연채광과 자연통풍을 효과적으로 활용할 수 있다.  

5. **녹화 및 수공간 활용**  
옥상녹화, 벽면녹화, 나무 식재 등을 통해 태양복사를 줄이고 주변 온도를 낮출 수 있다. 또한 연못이나 분수와 같은 수공간은 증발냉각 효과를 이용하여 쾌적한 환경을 조성한다.  

6. **지열 이용**  
지중의 온도가 연중 비교적 일정한 점을 이용하여 외기를 냉각하거나 예열할 수 있다. 여름에는 냉방효과를, 겨울에는 난방효과를 얻을 수 있다.  

7. **자연채광 활용**  
낮 동안 자연광을 최대한 이용하면 조명 사용을 줄일 수 있으며, 실내 열환경 개선에도 도움이 된다.  

이와 같은 방법들은 자연에너지를 이용하여 기계식 공조설비의 사용을 줄이고 에너지 소비를 절감할 수 있다는 장점이 있다. 따라서 공기조화에서는 기계 장치에만 의존하기보다 자연환기, 단열, 차양, 지열, 녹화 등을 함께 활용하는 것이 바람직하다.

### **Problem 04**  

기계적인 공기조화를 위해서는 온도와 습도, 기류, 청정도 등을 조절하여야 한다. 그리고 이들이 얼마나 정확하게 조절되고 있는지를 파악하기 위해서는 각종 계측기가 필요하다. 그중에서 공기조화에서 사용될 수 있는 것으로 온도와 습도를 측정할 수 있는 계측기로는 어떠한 것이 있는지 조사해 보자.

#### **Answer**

공기조화에서는 실내 환경을 쾌적하게 유지하기 위해 온도, 습도, 기류 및 청정도를 정확하게 조절하여야 한다. 이를 위해서는 각종 계측기를 사용하여 현재의 공기 상태를 측정하고 감시할 필요가 있다. 특히 온도와 습도는 공기조화에서 가장 기본적이고 중요한 요소이므로 다양한 측정기기가 사용된다.  

온도를 측정하는 계측기에는 다음과 같은 것들이 있다.

1. **유리온도계**  
가장 기본적인 온도계로서 액체의 열팽창 원리를 이용한다. 수은온도계와 알코올온도계가 대표적이다. 구조가 간단하고 사용이 쉬워 일반적으로 널리 사용된다.  

2. **바이메탈 온도계**  
열팽창 계수가 서로 다른 두 금속을 접합하여 만든 온도계이다. 온도 변화에 따라 금속이 휘어지는 원리를 이용하여 온도를 측정한다.  

3. **열전대(Thermocouple)**  
서로 다른 두 종류의 금속을 접합하였을 때 발생하는 열기전력을 이용하여 온도를 측정하는 장치이다. 측정 범위가 넓고 응답속도가 빠르다.  

4. **저항온도계(RTD)**  
금속의 전기저항이 온도에 따라 변하는 원리를 이용한 온도계이다. 정확도가 높아 공조설비와 산업설비에서 많이 사용된다.  

5. **서미스터(Thermistor)**  
반도체의 전기저항 변화 특성을 이용한 온도센서이다. 소형이며 민감도가 높아 전자식 공조기기에 많이 사용된다.  

습도를 측정하는 계측기에는 다음과 같은 것들이 있다.

1. **건습구 습도계**  
건구온도계와 습구온도계를 함께 사용하여 두 온도의 차이로 상대습도를 구하는 장치이다. 공기조화 분야에서 가장 기본적으로 사용된다.  

2. **모발 습도계**  
사람이나 동물의 모발이 습도에 따라 길이가 변하는 성질을 이용한 습도계이다. 구조가 간단하여 실내 습도 측정에 사용된다.  

3. **전기식 습도계**  
습도 변화에 따라 전기저항이나 정전용량이 변하는 원리를 이용한 습도계이다. 디지털 공조시스템에서 널리 사용된다.  

4. **노점계(Dew Point Meter)**  
공기 중 수증기가 응결하기 시작하는 온도인 노점을 측정하여 습도를 구하는 장치이다. 정밀한 습도 측정에 사용된다.  

5. **자동기록 온습도계**  
온도와 습도를 동시에 측정하고 기록할 수 있는 장치이다. 공조설비의 운전 상태를 지속적으로 관리하는 데 사용된다.  

이와 같은 계측기들은 공기조화 설비의 운전 상태를 정확하게 파악하고 쾌적한 실내 환경을 유지하기 위해 매우 중요하다. 최근에는 디지털 센서와 자동제어 기술의 발전으로 정밀하고 자동화된 계측 시스템이 널리 사용되고 있다.

### **Problem 05**  

국내·외에서 공기조화설비에 관한 정보를 제공하는 단체를 조사하고, 각 단체에서 제공하는 정보의 내용 및 각 단체의 회원으로 가입할 수 있는 조건 등에 대해 조사해 보자.

#### **Answer**

공기조화설비 분야에서는 기술 개발, 설계 기준 마련, 학술 연구 및 기술자 교육을 위해 다양한 전문 단체들이 활동하고 있다. 이러한 단체들은 공기조화 및 냉동 분야의 기술 정보, 설계 기준, 학술논문, 교육 프로그램 등을 제공하며 관련 기술 발전에 중요한 역할을 한다.  

국내·외의 대표적인 공기조화설비 관련 단체와 제공 정보 및 회원 가입 조건은 다음과 같다.


#### 1. 대한설비공학회(SAREK)

#### **Problem 06**  

국내·외에서 공기조화설비와 관련하여 출간되고 있는 서적으로 어떤 것이 있는지와 학교 도서관에 비치되어 있는 서적은 어떤 것이 있는지 조사해 보자.

#### **Answer**

공기조화설비분야에서는 냉난방, 환기, 냉동설비, 건축설비 및 에너지 절약 기술과 관련된 다양한 전문 서적이 국내·외에서 출간되고 있다. 이러한 서적들은 대학 교재, 기술자 실무서, 설계 참고서 및 연구 자료 등으로 널리 활용되고 있다.  

국내에서 출간되고 있는 대표적인 공기조화설비 관련 서적은 다음과 같다.

1. **공기조화시스템**  
공기조화설비의 기본 이론, 설계, 시공 및 유지관리 내용을 다룬 대표적인 전공 서적이다. 공조설비의 기초 이론과 실무 내용을 함께 설명하고 있어 대학 교재와 실무 참고서로 많이 사용된다. 

2. **공기조화냉동공학**  
냉동사이클, 공기조화 원리, 열전달 및 습공기선도 등을 설명하는 기초 전공 서적이다. 대학의 냉동공조 및 건축설비 관련 학과에서 주교재로 활용된다.  

3. **건축설비계획 및 설계**  
건축물의 냉난방, 환기, 위생설비 및 에너지 절약 설계를 종합적으로 설명하는 서적이다. 건축설비기사 및 공조냉동기계기사 시험 준비에도 활용된다.  

4. **냉동 및 공기조화**  
냉동기, 보일러, 냉각탑 및 공기조화기의 구조와 운전 원리를 설명한 실무 중심 서적이다.  

국외에서 출간되고 있는 대표적인 공기조화설비 관련 서적은 다음과 같다.

1. **ASHRAE Handbook**  
ASHRAE(American Society of Heating, Refrigerating and Air-Conditioning Engineers)에서 발간하는 가장 대표적인 HVAC 전문 서적이다. HVAC 설계 기준, 열부하 계산, 공조시스템 설계 및 실내공기질 등을 종합적으로 다루며 세계적으로 표준 참고서로 사용된다.

2. **Modern Refrigeration and Air Conditioning**  
냉동 및 공기조화의 기본 원리와 장비 운전, 유지관리 및 제어 기술을 설명하는 대표적인 교재이다. HVAC 교육기관과 기술학교에서 널리 사용된다.

3. **Refrigeration and Air Conditioning Technology**  
냉동 및 공기조화 시스템의 구성 요소와 고장 진단, 설계 및 운전 기술을 다룬 전문 서적이다. 실무 기술자와 학생들에게 많이 활용된다.  

4. **HVAC Handbook**  
HVAC 시스템의 설계와 시공, 열부하 계산 및 공기조화 이론을 종합적으로 정리한 참고서이다. 설계 엔지니어와 연구자들이 많이 활용한다.

학교 도서관에 일반적으로 비치되어 있는 공기조화설비 관련 서적으로는 다음과 같은 것들이 있다.

1. 공기조화냉동공학  
2. 건축설비계획  
3. 냉동 및 공기조화  
4. ASHRAE Handbook  
5. 열역학 및 열전달 교재  
6. 유체역학 및 공기조화 관련 참고서  

이러한 서적들은 공기조화설비의 기초 이론부터 설계 및 유지관리 기술까지 폭넓게 다루고 있으며, 대학 교육과 산업 현장에서 중요한 학습 자료로 활용되고 있다.

### **Problem 07**

국내에서 공기조화 기술자로 인정받을 수 있는 자격제도로는 어떤 것이 있는지, 그리고 자격을 취득하기 위해서는 어떻게 해야 하는지 등에 대해 조사해 보자.

#### **Answer**

국내에서 공기조화 기술자로 인정받을 수 있는 대표적인 자격제도로는 국가기술자격인 공조냉동기계 분야 자격증이 있다. 이러한 자격은 공기조화 및 냉동설비의 설계, 시공, 유지관리 및 안전관리를 수행할 수 있는 전문기술자를 양성하기 위해 운영되고 있다.  

대표적인 자격제도는 다음과 같다.

1. **공조냉동기계기능사**  
공기조화 및 냉동설비의 기초적인 운전과 유지관리 능력을 평가하는 자격이다. 현장 실무 중심의 기술 인력을 양성하기 위한 자격으로 냉동기 및 공조설비의 기본 작업을 수행할 수 있다.  

2. **공조냉동기계산업기사**  
공조냉동설비의 시공, 운전 및 관리 능력을 평가하는 자격이다. 기능사보다 높은 수준의 전문지식과 실무능력이 요구된다.  

3. **공조냉동기계기사**  
공조냉동설비의 설계, 시공, 유지관리 및 안전관리 등을 수행할 수 있는 전문기술자 자격이다. 건축설비 및 산업설비 분야에서 가장 대표적인 공조 관련 국가기술자격으로 인정받고 있다.

4. **공조냉동기계기술사**  
공조냉동 및 공기조화 분야의 최고 수준 국가기술자격이다. 설계, 감리, 기술지도 및 연구개발 등의 고급 기술업무를 수행할 수 있다. 풍부한 실무경험과 전문지식이 요구된다.  

이러한 자격증은 한국산업인력공단(Q-Net)에서 시행하는 국가기술자격시험을 통해 취득할 수 있다.  

자격을 취득하기 위한 방법은 다음과 같다.

1. **응시자격 충족**  
- 기능사는 특별한 제한 없이 응시 가능하다.  
- 산업기사와 기사는 관련 학과 졸업자 또는 일정 기간 이상의 실무경력이 필요하다.  
- 기술사는 기사 자격 취득 후 일정 기간 이상의 실무경력이 요구된다.  

2. **필기시험 합격**  
공기조화, 냉동공학, 열역학, 유체역학, 전기 및 자동제어 등 관련 이론 과목에 대한 시험을 치른다.  

3. **실기시험 합격**  
공조설비 설계, 냉동장치 운전, 배관작업 및 설비 유지관리 등 실무 능력을 평가한다.  

4. **자격증 취득 후 진출 분야**  
냉동 및 공조설비 업체, 건설회사, 엔지니어링 회사, 플랜트 및 산업설비 분야, 유지관리 업체, 공공기관 등 다양한 분야로 진출할 수 있다.

또한 최근에는 에너지 절약과 친환경 건축물의 중요성이 커지면서 공조냉동기계 분야 기술자의 수요도 증가하고 있다. 따라서 관련 자격증은 공기조화설비 분야의 전문성을 인정받는 중요한 기준이 되고 있다.
